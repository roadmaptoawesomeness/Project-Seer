```python
import pandas as pd
import numpy as np

import os

from time import time
from datetime import timedelta

import dask.dataframe as dd

pd.options.display.max_rows = 100
pd.set_option('display.float_format', '{:.0f}'.format)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
costs=dict()
avail=dict()
reload=dict()

start=time()
for i in os.listdir(path+'Prophet Test Thresh R/')[3:]:
    print(i)
    if ".csv" in i:
        df = pd.read_csv(path+'Prophet Test Thresh R/'+i).assign(file=i) 
        atm_no = np.array(['avail' in  x for x  in df.columns]).sum()
        costs[i] = df[['civ_cost','reload_cost','total_cost','cashout_flag','reload_flag']].sum()
        avail[i] = (df[['avail_' +str(i) for i in range(atm_no)]].sum()/df.shape[0]*100).astype(int).reset_index(drop=True)
        reload[i] = df[['start_reload_'+str(i) for i in range(atm_no)]].max().values
print(timedelta(seconds=time()-start))
```

```python
mavail = pd.DataFrame(avail).T.reset_index()
mavail['atm'] = mavail['index'].str.split("_").str[1]
mavail['model'] = mavail['index'].str.split("_").str[2]
mavail['atm_no'] = mavail.iloc[:,[1,2,3,4]].apply(lambda x: sum(x>0), axis=1)
```

```python
mcost = pd.DataFrame(costs).T.reset_index()
mcost['atm'] = mcost['index'].str.split("_").str[1]
mcost['model'] = mcost['index'].str.split("_").str[2]
mcost = mcost.groupby('atm').apply(pd.DataFrame.sort_values,['cashout_flag','total_cost']).reset_index(drop=True)
mcost = mcost.merge(mavail[['index','atm_no']],how='left')
mcost['reload_per_atm']=mcost.reload_flag/mcost.atm_no
best_cashout = mcost.drop_duplicates(subset=['atm']).copy()
```
